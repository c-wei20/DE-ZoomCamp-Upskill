```python
# get the DDL that pandas will use to CREATE table based on our connection (In this case is Postgres)
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# write the dataset to the DB
# with head(n=0) means only create the table with the column and without insert any data
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```
